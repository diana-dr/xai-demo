```python
temp = df_num.drop(columns = ['duration_ms', 'id', 'tempo', 'popularity', 'loudness', 'key', 'time_signature'])
diffs = {}
similar = {}
different = {}
for col in temp.columns:
    if (abs(df_num1[col] - df_num2[col]) <= 0.1).bool():
        similar[col] = abs(temp[col].diff()[1])
    else: different[col] = abs(temp[col].diff()[1])

    diffs[col] = abs(temp[col].diff()[1])
```

```python
similar = dict(sorted(similar.items(), key=lambda x:x[1]))
different = dict(sorted(different.items(), key=lambda x:x[1]))
```

```python
size = list(diffs.values())

fig = go.Figure(data=[go.Scatter(
    x=list(diffs.keys()),
    y=list(diffs.values()),
    text=['Difference' for i in diffs.keys()],
    mode='markers',
    marker=dict(
        size=size,
        sizemode='area',
        sizeref=1.*max(size)/(40.**2),
        sizemin=10
    )
)])

fig.update_layout(
    autosize=False,
    width=1000,
    height=600,)

fig.show()
```

```python
sizes = list(diffs.values())

fig = go.Figure(data=[go.Scatter(
    x=list(similar.keys()),
    y=list(similar.values()),
    text=['Difference' for i in similar.keys()],
    name='Features with Similar Values',
    mode='markers',
    marker=dict(
        size=sizes,
        sizemode='area',
        sizeref=0.2*max(sizes)/(40.**2),
        sizemin=10,
    )
), go.Scatter(
    x=list(different.keys()),
    y=list(different.values()),
    text=['Difference' for i in different.keys()],
    name='Features with Dissimilar Values',
    mode='markers',
    marker=dict(
        size=sizes,
        sizemode='area',
        sizeref=0.2*max(sizes)/(40.**2),
        sizemin=20,
    ),
)], layout=dict(
    legend=dict(
        itemsizing='constant'
    )
))

fig.update_layout(
    autosize=False,
    width=1000,
    height=600,)

fig.show()
```
